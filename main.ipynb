```python
import pandas as pd
import numpy as np
```

```python
head_data= ['size', 'bedroom', 'price']
data = pd.read_csv('home.txt', names=head_data, delimiter=',')
data
```

```python
def min_max_scaling(dataset):
    normalized_dataset = dataset.copy()
    
    for column in dataset.columns:
        min_value = dataset[column].min()
        max_value = dataset[column].max()
    
        normalized_dataset[column] = (dataset[column] - min_value) / (max_value - min_value)
    
    return normalized_dataset
```

I chose data normalization because the data has several outliers, for example the data in the first row has a value in the bedroom is 3 and the price is 399900, this quite high range of numbers made me decide to use normalization rather than standardization

```python
data = min_max_scaling(data)
data
```

```python
import pandas as pd

def train_test_split(data, test_size=0.2):
    # shuffled_data = data.sample(frac=1, random_state=42)
    
    #Sizing
    test_data_size = int(len(data) * test_size)

    train_data = data.iloc[:test_data_size]
    test_data = data.iloc[test_data_size:]

    
    # X_train y_train X_test y_test
    X_train = train_data[['size', 'bedroom']].values
    y_train = train_data['price'].values.reshape(-1, 1) 
    X_test = test_data[['size', 'bedroom']].values
    y_test = test_data['price'].values.reshape(-1, 1)
    
    return X_train, y_train, X_test, y_test
```

```python
data = pd.DataFrame(data, columns=['size', 'bedroom', 'price'])

X_train, y_train, X_test, y_test = train_test_split(data)

print("Training Set:")
print("X_train:", X_train)
print("y_train:", y_train)

print("\n\nTest Set:")
print("X_test:", X_test)
print("y_test:", y_test)
```

```python
class MultipleLinearRegression:
    def __init__(self):
        self.coeff_ = None
        self.intercept_ = None 

    def fit(self, X, y):
        X = np.insert(X, 0, 1, axis=1)
        
        betas = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
        self.intercept_ = betas[0]
        self.coeff_ = betas[1:]

    def predict(self, X):
        y_pred = np.dot(X, self.coeff_) + self.intercept_
        return y_pred
```

```python
mlr_model = MultipleLinearRegression()

mlr_model.fit(X_train, y_train)

y_pred = mlr_model.predict(X_test)

print("Training Set Predictions:")
print(y_pred)
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def r2_score(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    return r2


# Evaluasi model pada test set
test_rmse = rmse(y_test, y_pred)
test_mae = mae(y_test, y_pred)
test_r2 = r2_score(y_test, y_pred)

print("\nTest Set Evaluation:")
print("with normalization dataset and Multiple Linear Regression\n")
print("RMSE:", test_rmse)
print("MAE:", test_mae)
print("R^2:", test_r2)
```
